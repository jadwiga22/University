# Kwadratura Newtona-Cotesa

```julia
ENV["GKSwstype"] = "nul"
using QuadGK
using Plots
```

Funkcja <code>Integral</code> stanowi u nas wartość dokładną całki z danej funkcji.

```julia
function Integral( f, a, b )
    return quadgk( f, a, b )[1]
end
```

Funkcja <code>Trapezoid</code> jest kwadraturą złożoną korzystającą ze wzoru trapezów.

```julia
function Trapezoid( f, a, b, n )
    h = (b-a)/n
    X = collect(a : h : b)
    Y = f.(X)
    Y .*= h
    Y[1] *= 0.5
    Y[n+1] *= 0.5

    return sum(Y)
end
```

Funkcja <code>Simpson</code> jest kwadraturą złożoną korzystającą ze wzoru Simpsona. Argument *n*, który jej podajemy, powinien być liczbą parzystą. 

```julia
function Simpson( f, a, b, n )
    h = (b-a)/n
    X = collect(a : h : b)
    Y = f.(X)

    for i in 1:(n/2)-1
        Y[Int(2*i+1)] *= 2
    end

    for i in 1:(n/2)
        Y[Int(2*i)] *= 4
    end

    Y .*= (h/3)

    return sum(Y)
end
```

Funkcja <code>RelativeError</code> zwraca błąd względny przybliżenia *approx* wartości *x*.

```julia
function RelativeError( x, approx )
    return abs((x-approx)/x)
end
```

Funkcja <code>AbsoluteError</code> zwraca błąd bezwzględny przybliżenia *approx* wartości *x*.

```julia
function AbsoluteError( x, approx )
    return abs(x-approx)
end
```

Funkcja <code>IntegralError</code> zwraca tablicę $\frac{n}{2}$ wartości będących błędami (względnymi lub bezwzględnymi - w zależności od <code>ErrorType</code>) przybliżenia wartości całki z funkcji $f$ na przedziale $[a, b]$ przy użyciu metody <code>method</code> dla $2, 4, \ldots, n$ węzłów.

```julia
function IntegralError( f, a, b, n, method, ErrorType )
    ret = zeros(Float64, (Int(n/2),))
    for i in 2:2:n
        res = Integral( f, a, b )
        approx = method( f, a, b, i )
        ret[Int(i/2)] = ErrorType( res, approx )
    end
    return ret
end
```

Funkcja <code>DrawPlot</code> rysuje wykres błędów względnych metod trapezów i Simpsona dla $2, 4, \ldots, n$ węzłów.

```julia
function DrawPlot( f, a, b, n, ErrorType )
    if( ErrorType )
        error = RelativeError
    else
        error = AbsoluteError
    end
    retTrap = IntegralError( f, a, b, n, Trapezoid, error )
    retSimp = IntegralError( f, a, b, n, Simpson, error )
    plot(2:2:n, [retTrap retSimp], label=["Trapezoid" "Simpson"], lw=2)
end
```

```julia
function W3(x)
    res = 0
    for i in 1:1023
        res += (-1)^(i+1)*(1/i)*(x-1)^i
    end
    return res
end
DrawPlot(W3,-1,1,100,true)
```

## Pomiary czasu

Funkcja <code>CalculateWithTime</code> zwraca najmniejszy możliwy błąd względny przybliżenia całki z funkcji $f$ na przedziale $[a,b]$ przy użyciu metody <code>method</code>, przy ograniczeniu czasowym na wykonanie obliczeń równym $0,001$ sekundy.

```julia
function CalculateWithTime( method, f, a, b )
    exact_res = Integral(f, a, b)
    res = 1e18
    
    for i in 2:2:100
        ret = @elapsed method(f,a,b,i)
        cur = method(f,a,b,i)
        if(ret < 1e-3)
            res = min(res, RelativeError(exact_res,cur))
        end
    end
    
    return res
end
```

Funkcja <code>BestResWithTime</code> zwraca tablicę postaci $[x_1, x_2]$, gdzie $x_1$ jest wynikiem funkcji <code>CalculateWithTime</code> dla metody trapezów, a $x_2$ - dla metody Simpsona.

```julia
function BestResWithTime( f, a, b )
    return [CalculateWithTime(Trapezoid,f,a,b), CalculateWithTime(Simpson,f,a,b)]
end
```
